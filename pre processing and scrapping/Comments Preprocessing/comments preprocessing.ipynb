```python
array_of_csv=[#'Bioderma.csv',
              #'conc_mayb.csv',
              #'Era-ve-sunscreen.csv',
              #'found_mayb.csv',
              #'HP-envy-360.csv',
              #'lenovo-thinkpad-e14.csv',
              #'Lenovo-thinkpad-x.csv',
              #'Macbook-M1.csv',
              #'Neutrogena.csv',
             #'Macbook Air M2.csv',
             #'HP envy 13.csv',
             'galaxxy7']
array_of_txt=[#'Bioderma.txt',
              #'Conc_mayb.txt',
              #'Era ve sunscreen.txt',
              #'foun_mayb.txt',
              #'Hp envy 360.txt',
              #'Lenovo thinkpad e14.txt',
              #'Macbook M1.txt',
              #'Neutrogena.txt',
              #'Macbook Air M2.csv',
             #'HP envy 13.csv',
              'galaxy7.csv']
for i in range(len(array_of_txt)):
    
    with open(array_of_txt[i],'r',encoding='utf-8') as file:
            dataframe = file.read().replace('\n', '')
    lower_dataframe = dataframe.lower()
    import re
    NoNum_dataframe = re.sub(r'\d+', '', lower_dataframe)
    import string
    NoPunc_dataframe = re.sub(r'[^\w\s]', '', NoNum_dataframe)
    Cleaned_data = NoPunc_dataframe.strip()
    import nltk
    from nltk.corpus import stopwords 
    from nltk.tokenize import word_tokenize 
  
    stop_words = set(stopwords.words('english')) 
  
    word_tokens = word_tokenize(Cleaned_data) 
  
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
    filtered_sentence = [] 
  
    for w in word_tokens: 
         if w not in stop_words: 
            filtered_sentence.append(w) 

    #print(word_tokens) 
    #print(filtered_sentence) 
    import nltk
    from nltk.stem.porter import PorterStemmer
    from nltk.stem import WordNetLemmatizer

    stemmed_data=[]
    stemmer = PorterStemmer()
    for word in filtered_sentence:
     
          stemmed_data.append(stemmer.stem(word))
    #print(stemmed_data)
    lemitized_data = [] 
    lemmatizer = WordNetLemmatizer()
    for word in stemmed_data:

        lemitized_data.append(lemmatizer.lemmatize(word))
    #print(lemitized_data)
    import csv  
    with open(array_of_csv[i], 'w', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(lemitized_data)
    
```
