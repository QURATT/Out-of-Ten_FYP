# hp_envy

```python
import pandas as pd

dataframe = pd.DataFrame({'Sentiment name': ['Positive', 'Positive', 'Positive', 
                                   'Negative', 'Negative', 'Negative',
                                   'Neutral', 'Neutral', 'Neutral'],
                          'Sentiment': [11, 16, 328, 120,
                                                  6, 12, 8, 407, 
                                                  5]})
  
  

dataframe.groupby(['Sentiment name']).sum().plot(
    kind='pie', y= 'Sentiment', autopct='%1.1f%%', explode=[0.1,0.1,0.1],startangle=100,labeldistance=1.35,radius=1)
```
